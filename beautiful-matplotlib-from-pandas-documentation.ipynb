```python
#Source : https://pandas.pydata.org/pandas-docs/stable/visualization.html

# Related to iris.data

from pandas import read_csv
from pandas.tools.plotting import andrews_curves
import matplotlib.pyplot as plt

plt.style.use('ggplot')
data = read_csv('https://raw.githubusercontent.com/pydata/pandas/master/pandas/tests/data/iris.csv')
plt.figure(figsize=(14, 8))

andrews_curves(data, 'Name', colormap = 'winter')
```

```python
# Related to iris.data

from pandas.tools.plotting import parallel_coordinates
import matplotlib.pyplot as plt

plt.style.use('ggplot')
data = read_csv('https://raw.githubusercontent.com/pydata/pandas/master/pandas/tests/data/iris.csv')
plt.figure(figsize=(14, 8))

parallel_coordinates(data, 'Name')
```

```python
# Not related to iris.data

from pandas.tools.plotting import lag_plot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize = (14, 8))
data = pd.Series(0.1 * np.random.rand(1000) + 0.9 * np.sin(np.linspace(-99 * np.pi, 99 * np.pi, num=1000)))

lag_plot(data)
```
